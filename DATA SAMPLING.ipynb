```python
#data analysis library
import pandas as pd 
```

```python
#read trainset
train_data = pd.read_csv('Train_set.csv')
```

```python
#read testset
test_data = pd.read_csv('Test_set.csv')
```

- shape attribute - print the rows and columns, 
- shape[0] - print number of rows only
- shape[1] - print number of columns only

```python
#print function to echo the shape of the train_set data
print(train_data.shape[0],'rows and', train_data.shape[1], 'columns')
```

```python
#print function to echo the shape of the train_set data
print(test_data.shape[0],'rows and', train_data.shape[1], 'columns')
```

- sample() - sample function allows a subset of defined numbers of data to be selected randomly
- random_state - allow the sample data selected to be reproduceable. NB: the no can be any number, just to hold the state

```python
#select 100,000 random sample from trainset
sample_train = train_data.sample(100000, random_state = 50)

#select 50,000 random sample from testset
sample_test = test_data.sample(50000, random_state = 50)
```

```python
#print top 2 records
sample_train.head(2)
```

- reset_index() - reset_index function reset the unordered index number

```python
#reset the randomized sample trainset data index to avoid error for future use
sample_train = sample_train.reset_index(drop = True)

#reset the randomized sample testset data index to avoid error for future use
sample_test = sample_test.reset_index(drop = True)
```

```python
#print trainset & testset shape(rows and columns)
print (sample_train.shape, sample_test.shape)
```

```python
#print top 2 rows
sample_train.head(2)
```

```python
#print column names
sample_train.columns
```

```python
#print column names
sample_test.columns
```

# Save Sample Data as a new file

- to_csv() - a function that allow a data to be saved as a csv file
- index: index = False disallow the to_csv function not to create an index as a new column

```python
#save randomized sample of trainset as an excel file(csv format)
sample_train.to_csv('sample_train.csv', index=False)

#save randomized sample of testset as an excel file(csv format)
sample_test.to_csv('sample_test.csv',  index=False)
```
